```python
# Broader Cohort Groups by number of Months
```

```python
def cohort(tenure):
    if tenure < 13:
        return '0-12 Months'
    elif tenure < 25:
        return '12-24 Months'
    elif tenure < 49:
        return '24-48 Months'
    else:
        return 'over 48 Months'
```

```python
df['Tenure Cohort'] = df['tenure'].apply(cohort)
```

```python
df[['Tenure Cohort','tenure']]
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.scatterplot(df,x='MonthlyCharges',y='TotalCharges',hue='Tenure Cohort',alpha=0.4,linewidth=0.3,palette='Dark2')
```

```python
df_num = df.select_dtypes(include=[np.number])
```

```python
df_obj = df.select_dtypes(include='object')
```

```python
df_dummy = pd.get_dummies(df_obj,drop_first=True)
```

```python
fina_df = pd.concat([df_dummy,df_num],axis=1)
```

```python
fina_df
```

```python
fina_df = fina_df.astype(int)
```

```python
fina_df
```

```python
# Model Building
```

```python
X = fina_df.drop('Churn',axis=1)
```

```python
y = fina_df['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_x_train = scaler.fit_transform(X_train)
```

```python
scaled_x_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression(solver='saga',multi_class="ovr",max_iter=5000)
```

```python
penalty = ['l1','l2']
C = np.logspace(0,4,10)
```

```python
C
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(log_model,param_grid={'C':C,'penalty':penalty})
```

```python
grid_model.fit(scaled_x_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
y_pred = grid_model.predict(scaled_x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
